```python
import matplotlib.pyplot as plt
import csv
import numpy as np
np.set_printoptions(threshold=np.nan)
from scipy import stats
results = []
resi = []

with open("Raw_confocal_data_standard_sample.csv") as csvfile:
    reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC) # change contents to floats
        
    for row in reader: # each row is a list
         results.append(row)
```

```python
for i in results:
   resi.append(i[0])
```

```python

resi = np.array(resi)
print(resi)
n = len(resi)/5
print(n)
X = np.arange(1,(len(resi)/5) + 1)
print(X)
```

```python
k=0
sui=0
while(k<len(resi)):
    array = resi[k:k+n]
    m = ((n*(np.sum(X*array))) - ((np.sum(X)) * (np.sum(array)))) / ((n * (np.sum(X*X))) - ((np.sum(X)) * (np.sum(X))))
    c = (array) - np.multiply(m,X)
    c=abs(c)
    sumabs = (np.sum(c) / n)
    avgofc = c - sumabs
    avgofc = abs(avgofc)
    ki = np.sum(avgofc)
    sui = sui + (ki/n)
    k = k+n
print(sui/5)    
    
        
    
```
